```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load CSV
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Revenue'] = df['Units Sold'] * df['Price']
df[['Product', 'Revenue']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(data=df, x='Product', y='Revenue')
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert date
sales_by_date = df.groupby('Date')['Revenue'].sum().reset_index()

plt.plot(sales_by_date['Date'], sales_by_date['Revenue'], marker='o')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
